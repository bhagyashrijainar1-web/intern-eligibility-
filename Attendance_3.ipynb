```python
# Attendance Summary Generator

n = int(input("Enter number of interns: "))

present = 0
absent = 0

for i in range(1, n + 1):
    status = input(f"Enter attendance for Intern {i} (P/A): ").upper()

    if status == "P":
        present += 1
    elif status == "A":
        absent += 1
    else:
        print("Invalid input! Counted as Absent.")
        absent += 1

print("\nAttendance Summary")
print("Total Present:", present)
print("Total Absent:", absent)
```
